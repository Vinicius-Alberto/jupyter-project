<H2>DistribuiÃ§Ã£o de vendas por regiÃ£o</H2>

```python
# Somando as vendas por regiÃ£o
region_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

# Criando o grÃ¡fico de anel
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', 
                                  startangle=140, colors=['red', 'blue', 'green', 'orange'], 
                                  wedgeprops=dict(width=0.4))  # Define a largura do anel

# Adicionando um cÃ­rculo branco no centro
plt.gca().add_artist(plt.Circle((0, 0), 0.3, color='white'))

# Ajustando a posiÃ§Ã£o dos nÃºmeros para ficarem dentro do anel
for autotext in autotexts:
    autotext.set_color('white')  # Define a cor do texto
    autotext.set_fontsize(12)     # Ajusta o tamanho da fonte
    autotext.set_weight("bold")   # Deixa o texto em negrito
    autotext.set_bbox(dict(facecolor='black', edgecolor='none', boxstyle='round,pad=0.2'))  # Adiciona um fundo escuro

plt.title("DistribuiÃ§Ã£o de Vendas por RegiÃ£o")
plt.show()
```
